## Load Libraries

```

import pandas as pd
import sqlite3
```

## Load Datasets

```

orders = pd.read_csv('data/orders.csv')
users = pd.read_json('data/users.json')

conn = sqlite3.connect(':memory:')
with open('data/restaurants.sql','r') as f:
    conn.executescript(f.read())
restaurants = pd.read_sql('SELECT * FROM restaurants', conn)
```

## Merge Datasets

```

merged = orders.merge(users, on='user_id', how='left')                .merge(restaurants, on='restaurant_id', how='left')
merged.head()
```

## Save Final Dataset

```

merged.to_csv('outputs/final_food_delivery_dataset.csv', index=False)
```

## Sample Analysis

```

merged.groupby('membership')['total_amount'].mean()
```
